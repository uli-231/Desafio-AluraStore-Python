### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
'''En este primer análisis, debes calcular el ingreso total de cada tienda.
Sumando los valores de la columna Precio de cada conjunto de datos de la tienda
 para estimar los ingresos.'''

def sumatoria (tienda,campo):
  return tienda[campo].sum()
```

```python
print(sumatoria(tienda,"Precio"))
print(sumatoria(tienda2,"Precio"))
print(sumatoria(tienda3,"Precio"))
print(sumatoria(tienda4,"Precio"))
```

# 2. Ventas por categoría

# En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
''' Definición de función para la creación de un diccionario con todas las categorías y conteo de las ventas de cada categoría'''

def conteo (tienda,campo):
  cant_ventas_categorias = {}
  for i in range(len(tienda[campo])):
    if tienda[campo][i] not in cant_ventas_categorias:
      cant_ventas_categorias.update({tienda[campo][i]:1})
    else:
      cant_ventas_categorias[tienda[campo][i]] += 1
  cant_ventas_categorias = dict(sorted(cant_ventas_categorias.items(), key=lambda item: item[1], reverse=True))
  return cant_ventas_categorias
```

```python
''' Aplicamos la función a cada una de las tiendas y mostramos los diccionarios'''

Categorias_tienda_1 = conteo(tienda,"Categoría del Producto")
Categorias_tienda_2 = conteo(tienda2,"Categoría del Producto")
Categorias_tienda_3 = conteo(tienda3,"Categoría del Producto")
Categorias_tienda_4 = conteo(tienda4,"Categoría del Producto")

print(Categorias_tienda_1)
print(Categorias_tienda_2)
print(Categorias_tienda_3)
print(Categorias_tienda_4)
```

```python
''' Mostramos la categoría con más ventas de cada tienda y la cantidad vendida'''

print(f'La categoría más vendida en la tienda 1 es: {list(Categorias_tienda_1.keys())[0]} con {list(Categorias_tienda_1.values())[0]} ventas')
print(f'La categoría más vendida en la tienda 2 es: {list(Categorias_tienda_2.keys())[0]} con {list(Categorias_tienda_2.values())[0]} ventas')
print(f'La categoría más vendida en la tienda 3 es: {list(Categorias_tienda_3.keys())[0]} con {list(Categorias_tienda_3.values())[0]} ventas')
print(f'La categoría más vendida en la tienda 4 es: {list(Categorias_tienda_4.keys())[0]} con {list(Categorias_tienda_4.values())[0]} ventas')
```

# 3. Calificación promedio de la tienda


# En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es **conocer la satisfacción del cliente con los productos vendidos**.

```python
''' Definición de una función donde se calculan los promedios de las calificaciones de cada tienda'''

def promedio (tienda,campo):
  ''' Función para calcular el promedio de un campo de un dataframe
  tienda: dataframe a analizar
  campo: campo a analizar
  '''
  return tienda[campo].mean()
```

```python
''' Aplicamos la función a cada una de las tiendas y mostramos los promedios redondeados a dos cifras'''

promedio_tienda_1 = promedio(tienda,"Calificación")
promedio_tienda_2 = promedio(tienda2,"Calificación")
promedio_tienda_3 = promedio(tienda3,"Calificación")
promedio_tienda_4 = promedio(tienda4,"Calificación")

print(f'La calificación promedio de la tienda 1 es: {round(promedio_tienda_1,2)}')
print(f'La calificación promedio de la tienda 2 es: {round(promedio_tienda_2,2)}')
print(f'La calificación promedio de la tienda 3 es: {round(promedio_tienda_3,2)}')
print(f'La calificación promedio de la tienda 4 es: {round(promedio_tienda_4,2)}')
```

# 4. Productos más y menos vendidos

#En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
''' Definición de una función donde se calculan las ventas por producto de cada tienda'''

def ventas_por_producto (tienda,campo):
  ''' Función para calcular las ventas por productos de una tienda
  tienda: dataframe a analizar
  campo: campo a analizar
  '''
  cant_ventas_productos = {}
  for i in range(len(tienda[campo])):
    if tienda[campo][i] not in cant_ventas_productos:
      cant_ventas_productos.update({tienda[campo][i]:1})
    else:
      cant_ventas_productos[tienda[campo][i]] += 1
  cant_ventas_productos = dict(sorted(cant_ventas_productos.items(), key=lambda item: item[1], reverse=True))
  return cant_ventas_productos
```

```python
''' Aplicamos la función a cada una de las tiendas y mostramos los resultados'''

ventas_por_producto_tienda_1 = ventas_por_producto(tienda,"Producto")
ventas_por_producto_tienda_2 = ventas_por_producto(tienda2,"Producto")
ventas_por_producto_tienda_3 = ventas_por_producto(tienda3,"Producto")
ventas_por_producto_tienda_4 = ventas_por_producto(tienda4,"Producto")

print(ventas_por_producto_tienda_1)
print(ventas_por_producto_tienda_2)
print(ventas_por_producto_tienda_3)
print(ventas_por_producto_tienda_4)
```

```python
'''Definimos una función para encontrar y almacenar en un diccionario el o los productos más vendidos y los menos vendidos de cada tienda'''

def productos_mas_menos_vendidos (lista_ventas_producto):
  ''' Función para encontrar y almacenar en un diccionario el o los productos más vendidos y los menos vendidos de una tienda
  lista_ventas_producto: diccionario a analizar
  '''
  cant_mas_ventas_productos = {}
  cant_menos_ventas_productos = {}
  for i in range(len(lista_ventas_producto)):
    if list(lista_ventas_producto.values())[i] == max(lista_ventas_producto.values()):
      cant_mas_ventas_productos.update({list(lista_ventas_producto.keys())[i]:list(lista_ventas_producto.values())[i]})
    elif list(lista_ventas_producto.values())[i] == min(lista_ventas_producto.values()):
      cant_menos_ventas_productos.update({list(lista_ventas_producto.keys())[i]:list(lista_ventas_producto.values())[i]})
  return cant_mas_ventas_productos, cant_menos_ventas_productos
```

```python
''' Aplicamos la función definida anteriormente y mostramos el o los productos más vendidos y los menos vendidos de cada tienda y la cantidad vendida de cada uno de ellos'''

mas_menos_vendidos_tienda_1 = productos_mas_menos_vendidos(ventas_por_producto_tienda_1)
mas_menos_vendidos_tienda_2 = productos_mas_menos_vendidos(ventas_por_producto_tienda_2)
mas_menos_vendidos_tienda_3 = productos_mas_menos_vendidos(ventas_por_producto_tienda_3)
mas_menos_vendidos_tienda_4 = productos_mas_menos_vendidos(ventas_por_producto_tienda_4)

print(mas_menos_vendidos_tienda_1)
print(mas_menos_vendidos_tienda_2)
print(mas_menos_vendidos_tienda_3)
print(mas_menos_vendidos_tienda_4)
```

```python
''' Mostramos los productos más vendidos y los menos vendidos de cada tienda'''

print(f'El/Los producto más vendido/s en la tienda 1 es/son: {list(mas_menos_vendidos_tienda_1[0].keys())} con {list(mas_menos_vendidos_tienda_1[0].values())[0]} ventas')
print(f'El/Los producto más vendido/s en la tienda 2 es/son: {list(mas_menos_vendidos_tienda_2[0].keys())} con {list(mas_menos_vendidos_tienda_2[0].values())[0]} ventas')
print(f'El/Los producto más vendido/s en la tienda 3 es/son: {list(mas_menos_vendidos_tienda_3[0].keys())} con {list(mas_menos_vendidos_tienda_3[0].values())[0]} ventas')
print(f'El/Los producto más vendido/s en la tienda 4 es/son: {list(mas_menos_vendidos_tienda_4[0].keys())} con {list(mas_menos_vendidos_tienda_4[0].values())[0]} ventas')

print(f'El/Los producto menos vendido/s en la tienda 1 es/son: {list(mas_menos_vendidos_tienda_1[1].keys())} con {list(mas_menos_vendidos_tienda_1[1].values())[0]} ventas')
print(f'El/Los producto menos vendido/s en la tienda 2 es/son: {list(mas_menos_vendidos_tienda_2[1].keys())} con {list(mas_menos_vendidos_tienda_2[1].values())[0]} ventas')
print(f'El/Los producto menos vendido/s en la tienda 3 es/son: {list(mas_menos_vendidos_tienda_3[1].keys())} con {list(mas_menos_vendidos_tienda_3[1].values())[0]} ventas')
print(f'El/Los producto menos vendido/s en la tienda 4 es/son: {list(mas_menos_vendidos_tienda_4[1].keys())} con {list(mas_menos_vendidos_tienda_4[1].values())[0]} ventas')
```

# 5. Envío promedio por tienda

#En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.

```python
''' Definición de una función donde se calculan los costos promedio de envío de cada tienda'''

def promedio_envio (tienda,campo):
  ''' Función para calcular el promedio de un campo de un dataframe
  tienda: dataframe a analizar
  campo: campo a analizar
  '''
  return tienda[campo].mean()
```

```python
''' Aplicamos la función a cada una de las tiendas y mostramos los promedios redondeados a dos cifras'''
promedio_envio_tienda_1 = promedio_envio(tienda,"Costo de envío")
promedio_envio_tienda_2 = promedio_envio(tienda2,"Costo de envío")
promedio_envio_tienda_3 = promedio_envio(tienda3,"Costo de envío")
promedio_envio_tienda_4 = promedio_envio(tienda4,"Costo de envío")

print(f'El costo promedio de envío de la tienda 1 es: $ {round(promedio_envio_tienda_1,2)}')
print(f'El costo promedio de envío de la tienda 2 es: $ {round(promedio_envio_tienda_2,2)}')
print(f'El costo promedio de envío de la tienda 3 es: $ {round(promedio_envio_tienda_3,2)}')
print(f'El costo promedio de envío de la tienda 4 es: $ {round(promedio_envio_tienda_4,2)}')
```
